```python
import pandas as pd
ratings = pd.read_csv('item_based.csv', header=0)
ratings.head()
```

```python
pivotedRatings = ratings.pivot_table(index=['staff_id'],columns=['candidate_id'],values='ratings (1-5)')
pivotedRatings.head()
```

```python
candidateRatings = pivotedRatings[3]
candidateRatings
```

```python
similarCandidates = pivotedRatings.corrwith(candidateRatings)
similarCandidates = similarCandidates.dropna()
df = pd.DataFrame(similarCandidates)
df.head(10)
```

```python
import numpy as np
candidateStats = ratings.groupby('candidate_id').agg({'ratings (1-5)': [np.size, np.mean]})
candidateStats.head()
```

```python
popularCandidates = candidateStats['ratings (1-5)']['size'] >= 2
candidateStats[popularCandidates].sort_values([('ratings (1-5)', 'mean')], ascending=False)[:15]
```

```python
df = candidateStats[popularCandidates].join(pd.DataFrame(similarCandidates, columns=['similarity']))
```

```python
df.sort_values(['similarity'], ascending=False)[1:15]
```
