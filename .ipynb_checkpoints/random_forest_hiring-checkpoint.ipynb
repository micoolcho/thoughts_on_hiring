```python
import numpy as np
import pandas as pd
from sklearn import tree

dummy_file = "dummyhires.csv"
dummy_hires = pd.read_csv(dummy_file, header = 0)
```

```python
dummy_hires.head()
```

Clean up some data into numerical values.

```python
d = {'Y': 1, 'N': 0}
dummy_hires['Hired'] = dummy_hires['Hired'].map(d)
dummy_hires['Top-Tier school'] = dummy_hires['Top-Tier school'].map(d)
d = {'BS': 0, 'MS': 1, 'PhD': 2}
dummy_hires['Education'] = dummy_hires['Education'].map(d)
dummy_hires.head()
```

```python
dummy_features = list(dummy_hires.columns[:6])
dummy_features
```

```python
y = dummy_hires["Hired"]
X = dummy_hires[dummy_features]
classifier = tree.DecisionTreeClassifier()
classifier = classifier.fit(X,y)
```

```python
from IPython.display import Image
from sklearn.externals.six import StringIO  
import pydot 

dot_data = StringIO()  
tree.export_graphviz(classifier, out_file=dot_data, feature_names=dummy_features)  
graph = pydot.graph_from_dot_data(dot_data.getvalue())  
Image(graph.create_png())  
```

Here's how we interpret the results above. A value of [2,0] means 2 are NOT hired and 0 are hired under the conditions at that node. At the top node, we see that 2 people are not hired and 0 are hired based on the 2 samples of candidates with interview ratings less than 2.5 (going left means the "true" branch of the condition above; interview ratings <=2.5).
